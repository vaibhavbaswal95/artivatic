```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

pd.set_option('display.max_columns', 60)
df = pd.read_csv('./ML_Artivatic_dataset/train_indessa.csv', )
df.head(5)
```

```python
##Finding the the count and percentage of values that are missing in the dataframe.
df_null = pd.DataFrame({'Count': df.isnull().sum(), 'Percent': 100*df.isnull().sum()/len(df)})

##printing columns with null count more than 0
df_null[df_null['Count'] > 0] 
```

```python
num_cols = list(df.describe().columns)
num_cols.remove('member_id')
num_cols
```

```python
df1 = df[num_cols]
df1 = df1.fillna(0)
```

```python
X = df1.loc[:, df1.columns != "loan_status"]
y = df1.loc[:, "loan_status"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
rf = RandomForestClassifier(
    n_estimators=200,
    criterion="gini",
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features="auto",
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    min_impurity_split=None,
    bootstrap=True,
    oob_score=False,
    n_jobs=1,
    random_state=None,
    verbose=0,
    warm_start=False,
    class_weight=None,
)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
fi = rf.feature_importances_

feature_cols = X_test.columns
importance = pd.DataFrame(
    {"feature": feature_cols, "importance": rf.feature_importances_}
)

importance = pd.DataFrame(
    {"feature": feature_cols[:], "importance": rf.feature_importances_[:]}
)
importance.sort_values(
    by="importance",
    axis=0,
    ascending=False,
    inplace=True,
    kind="quicksort",
    na_position="last",
)
importance.plot(x="feature", y="importance", kind="bar")
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

cols = list(importance.feature[:10])
cols.insert(0, "loan_status")
corrcoef_map = np.corrcoef(df1[cols].values.T)
fig, ax = plt.subplots(figsize=(20,12))  # Sample figsize in inches
hm = sns.heatmap(
    corrcoef_map,
    cbar=True,
    annot=True,
    square=True,
    fmt=".2f",
    annot_kws={"size": 15},
    yticklabels=cols,
    xticklabels=cols,
    ax=ax,
)
```
